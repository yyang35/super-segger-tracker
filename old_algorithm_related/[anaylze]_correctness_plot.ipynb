```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have already defined your DataFrame 'df'

# Calculate the average
average = df.mean()

# Create the bar plot
ax = average.plot(kind='bar', figsize=(10, 6))

# Add labels to the bars
for i, v in enumerate(average):
    ax.text(i, v + 0.01, str(round(v, 4)), ha='center', va='bottom')

# Set labels and title (optional)
ax.set_xlabel('Categories')
ax.set_ylabel('Average Value')
ax.set_title('Average Values Bar Chart')
plt.xticks(rotation=30, ha="right")

# Show the plot
plt.show()
```
